```python
import pandas as pd
import matplotlib.pyplot as plt

results_path_mm = "/content/drive/MyDrive/Semantics/dataset/frasi_bias/frasi_bias_mm.csv"
results_path_fm = "/content/drive/MyDrive/Semantics/dataset/frasi_bias/frasi_bias_fm.csv"
results_path_mf = "/content/drive/MyDrive/Semantics/dataset/frasi_bias/frasi_bias_mf.csv"
results_path_ff = "/content/drive/MyDrive/Semantics/dataset/frasi_bias/frasi_bias_ff.csv"
```

```python
mm_df = pd.read_csv(results_path_mm)
fm_df = pd.read_csv(results_path_fm)
mf_df = pd.read_csv(results_path_mf)
ff_df = pd.read_csv(results_path_ff)
```

```python
def print_percentages(df):
  df.columns = mm_df.columns.str.strip()
  category_percentages = df['categoria'].value_counts(normalize=True) * 100

  return category_percentages

def plot_percentages(category_percentages):
  plt.figure(figsize=(10, 6))
  category_percentages.plot(kind='bar', color='skyblue')
  plt.title('Category Percentages')
  plt.xlabel('Category')
  plt.ylabel('Percentage')
  plt.xticks(rotation=0)
  plt.show()
```

```python
category_percentages_mm = print_percentages(mm_df)
category_percentages_fm = print_percentages(fm_df)
category_percentages_mf = print_percentages(mf_df)
category_percentages_ff = print_percentages(ff_df)
```

```python
print("Percentuale di argomenti tra uomo e uomo")
print(category_percentages_mm)
plot_percentages(category_percentages_mm)
```

```python
print("Percentuale di argomenti tra uomo e donna")
print(category_percentages_mf)
plot_percentages(category_percentages_mf)
```

```python
print("Percentuale di argomenti tra donna e uomo")
print(category_percentages_fm)
plot_percentages(category_percentages_fm)
```

```python
print("Percentuale di argomenti tra donna e donna")
print(category_percentages_ff)
plot_percentages(category_percentages_ff)
```
